```python
import re
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, classification_report
```

```python
# Data
data = pd.read_excel('/content/data100.xlsx')
data = data.fillna('')
```

```python
# pre procesing

def preprocess_text(text):
    if isinstance(text, str):
        text = re.sub(r'\d+', '', text)
        text = re.sub(r'[^\w\s]', '', text)
        text = text.lower()
    return text

for key in data:
    data[key] = data[key].apply(preprocess_text)
# Preprocessing

```

```python
for label, documents in data.items():
    X.extend(documents)
    y.extend([label] * len(documents))

# Splitting data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=42)

# Vectorizing the text data (you can choose any other method as per your requirement)
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer()
X_train_vec = vectorizer.fit_transform(X_train)
X_test_vec = vectorizer.transform(X_test)

# Training a logistic regression model
model = LogisticRegression()
model.fit(X_train_vec, y_train)

# Predicting the classes for test data
y_pred = model.predict(X_test_vec)

# Creating a confusion matrix
cm = confusion_matrix(y_test, y_pred)

print("Confusion Matrix:")
print(cm)
```

```python
report = classification_report(y_test, y_pred)

print("Classification Report:")
print(report)
```
